## Candidate Matching

```python
query = """
// Candidate Matching by Campaign ID
MATCH (campaign:Campaign {id: $campaignId})
MATCH (campaign)-[:FOR_CAMPAIGN]->(job:Job)

// Join related info
OPTIONAL MATCH (job)-[:REQUIRES_SKILL]->(jobSkill:Skill)
OPTIONAL MATCH (candidate)-[:HAS_SKILL]->(candidateSkill:Skill)
OPTIONAL MATCH (job)-[:SPEAKS_LANGUAGE]->(jobLang:Language)
OPTIONAL MATCH (candidate)-[:SPEAKS_LANGUAGE]->(candidateLang:Language)
OPTIONAL MATCH (job)-[:HAS_REQUIREMENT]->(jobReq:Requirement)
OPTIONAL MATCH (candidate)-[:HAS_SPEC]->(candidateSpec:JobSpec)
OPTIONAL MATCH (candidate)-[:HAS_EXPERIENCE]->(exp:Experience)
OPTIONAL MATCH (candidate)-[:ATTENDED]->(edu:University)
OPTIONAL MATCH (candidate)-[:WORKED_AT]->(expCompany:Company)

// Collect data
WITH job, candidate,
     COLLECT(DISTINCT jobSkill) AS jobSkills,
     COLLECT(DISTINCT candidateSkill) AS candidateSkills,
     COLLECT(DISTINCT jobLang) AS jobLanguages,
     COLLECT(DISTINCT candidateLang) AS candidateLanguages,
     COLLECT(DISTINCT jobReq) AS jobRequirements,
     COLLECT(DISTINCT candidateSpec) AS candidateSpecs,
     COLLECT(DISTINCT exp) AS experiences,
     COLLECT(DISTINCT edu) AS education,
     COLLECT(DISTINCT expCompany) AS workHistory

// Compute score components
WITH job, candidate,
     jobSkills, candidateSkills, jobLanguages, candidateLanguages,
     jobRequirements, candidateSpecs, experiences, education, workHistory,

     // Skills match
     CASE WHEN SIZE(jobSkills) = 0 THEN 0 ELSE
         REDUCE(score = 0, js IN jobSkills |
             score + CASE
                 WHEN ANY(cs IN candidateSkills WHERE cs.name = js.name)
                 THEN CASE js.mastery
                     WHEN 'expert' THEN 10
                     WHEN 'advanced' THEN 8
                     WHEN 'intermediate' THEN 6
                     WHEN 'beginner' THEN 4
                     ELSE 5 END
                 ELSE 0
             END
         ) * 1.0 / SIZE(jobSkills)
     END AS skillsScore,

     // Language match
     CASE WHEN SIZE(jobLanguages) = 0 THEN 10 ELSE
         REDUCE(score = 0, jl IN jobLanguages |
             score + CASE
                 WHEN ANY(cl IN candidateLanguages WHERE cl.name = jl.name)
                 THEN CASE jl.mastery
                     WHEN 'native' THEN 10
                     WHEN 'fluent' THEN 8
                     WHEN 'intermediate' THEN 6
                     WHEN 'basic' THEN 4
                     ELSE 5 END
                 ELSE 0
             END
         ) * 1.0 / SIZE(jobLanguages)
     END AS languageScore,

     // Seniority
     CASE
         WHEN job.seniority_level IS NULL THEN 7
         WHEN job.seniority_level = candidate.seniority_level THEN 10
         WHEN job.seniority_level = 'senior' AND candidate.seniority_level = 'mid' THEN 8
         WHEN job.seniority_level = 'mid' AND candidate.seniority_level = 'junior' THEN 6
         ELSE 3
     END AS seniorityScore,

     // Experience relevance
     CASE WHEN SIZE(experiences) = 0 THEN 0 ELSE
         REDUCE(score = 0, e IN experiences |
             score + CASE
                 WHEN e.company_name = job.company_name THEN 8
                 WHEN e.role CONTAINS job.title THEN 6
                 WHEN e.industry = job.industry_sector THEN 4
                 ELSE 2 END
         ) * 1.0 / SIZE(experiences)
     END AS experienceScore,

     // Education level
     CASE WHEN SIZE(education) = 0 THEN 5 ELSE
         REDUCE(score = 0, ed IN education |
             score + CASE ed.degree
                 WHEN 'PhD' THEN 10
                 WHEN 'Masters' THEN 8
                 WHEN 'Bachelors' THEN 6
                 WHEN 'Associates' THEN 4
                 ELSE 3 END
         ) * 1.0 / SIZE(education)
     END AS educationScore,

     // Location match
     CASE
         WHEN candidate.location = job.location THEN 10
         WHEN candidate.remote_availability = true AND job.work_mode = 'remote' THEN 9
         WHEN candidate.remote_availability = true AND job.work_mode = 'hybrid' THEN 7
         WHEN candidate.location CONTAINS job.location OR job.location CONTAINS candidate.location THEN 6
         ELSE 3
     END AS locationScore,

     // Availability bonus
     CASE
         WHEN candidate.availability = 'immediate' THEN 2
         WHEN candidate.availability = 'two_weeks' THEN 1
         ELSE 0
     END AS availabilityBonus,

     // Salary alignment
     CASE
         WHEN job.salary_min IS NOT NULL AND job.salary_max IS NOT NULL AND candidate.salary_expectation IS NOT NULL THEN
             CASE
                 WHEN candidate.salary_expectation >= job.salary_min AND candidate.salary_expectation <= job.salary_max THEN 2
                 WHEN candidate.salary_expectation < job.salary_min THEN 1
                 ELSE -1
             END
         ELSE 0
     END AS salaryAlignmentBonus

// Total score
WITH candidate,
     (skillsScore * 0.35 +
      languageScore * 0.10 +
      seniorityScore * 0.20 +
      experienceScore * 0.15 +
      educationScore * 0.10 +
      locationScore * 0.10) +
     availabilityBonus + salaryAlignmentBonus AS totalScore,

     skillsScore, languageScore, seniorityScore, experienceScore, educationScore,
     locationScore, availabilityBonus, salaryAlignmentBonus

// Final output
RETURN 
  candidate.name AS candidateName,
  candidate.email AS candidateEmail,
  candidate.location AS candidateLocation,
  candidate.seniority_level AS candidateSeniority,
  candidate.availability AS candidateAvailability,
  candidate.salary_expectation AS salaryExpectation,
  ROUND(totalScore * 100)/100 AS totalScore,
  ROUND(skillsScore * 100)/100 AS skillsScore,
  ROUND(languageScore * 100)/100 AS languageScore,
  ROUND(seniorityScore * 100)/100 AS seniorityScore,
  ROUND(experienceScore * 100)/100 AS experienceScore,
  ROUND(educationScore * 100)/100 AS educationScore,
  ROUND(locationScore * 100)/100 AS locationScore,
  availabilityBonus,
  salaryAlignmentBonus

ORDER BY totalScore DESC
LIMIT 50;
"""

kg.query(query, params={"campaignId": "9b76cec9-8006-447c-9bae-26815c850b94"})
```

```python
kg.query("""
// Comprehensive Job-Candidate Ranking Query
// Replace 'target_job_title', 'target_company', 'target_location' with actual values

MATCH (job:Job {title: 'machine learning engineer'})

// Get all candidates and calculate comprehensive scoring
MATCH (candidate:Candidate)

// Optional matches for different scoring components
OPTIONAL MATCH (job)-[:REQUIRES_SKILL]->(jobSkill:Skill)
OPTIONAL MATCH (candidate)-[:HAS_SKILL]->(candidateSkill:Skill)
OPTIONAL MATCH (job)-[:SPEAKS_LANGUAGE]->(jobLang:Language)
OPTIONAL MATCH (candidate)-[:SPEAKS_LANGUAGE]->(candidateLang:Language)
OPTIONAL MATCH (job)-[:HAS_REQUIREMENT]->(jobReq:Requirement)
OPTIONAL MATCH (candidate)-[:HAS_SPEC]->(candidateSpec:JobSpec)
OPTIONAL MATCH (candidate)-[:HAS_EXPERIENCE]->(exp:Experience)
OPTIONAL MATCH (candidate)-[:ATTENDED]->(edu:University)
OPTIONAL MATCH (candidate)-[:WORKED_AT]->(company:Company)

WITH job, candidate, 
     // Collect all related data for scoring
     COLLECT(DISTINCT jobSkill) AS jobSkills,
     COLLECT(DISTINCT candidateSkill) AS candidateSkills,
     COLLECT(DISTINCT jobLang) AS jobLanguages,
     COLLECT(DISTINCT candidateLang) AS candidateLanguages,
     COLLECT(DISTINCT jobReq) AS jobRequirements,
     COLLECT(DISTINCT candidateSpec) AS candidateSpecs,
     COLLECT(DISTINCT exp) AS experiences,
     COLLECT(DISTINCT edu) AS education,
     COLLECT(DISTINCT company) AS workHistory

// Calculate individual scoring components
WITH job, candidate,
     jobSkills, candidateSkills, jobLanguages, candidateLanguages,
     jobRequirements, candidateSpecs, experiences, education, workHistory,
     
     // 1. Skills Match Score (Weight: 35%)
     CASE 
       WHEN SIZE(jobSkills) = 0 THEN 0
       ELSE REDUCE(score = 0, jobSkill IN jobSkills | 
         score + CASE 
           WHEN ANY(candSkill IN candidateSkills WHERE candSkill.name = jobSkill.name) 
           THEN 
             CASE jobSkill.mastery
               WHEN 'expert' THEN 10
               WHEN 'advanced' THEN 8
               WHEN 'intermediate' THEN 6
               WHEN 'beginner' THEN 4
               ELSE 5
             END
           ELSE 0
         END) * 1.0 / SIZE(jobSkills)
     END AS skillsScore,
     
     // 2. Language Match Score (Weight: 10%)
     CASE 
       WHEN SIZE(jobLanguages) = 0 THEN 10 // Full score if no language requirements
       ELSE REDUCE(score = 0, jobLang IN jobLanguages | 
         score + CASE 
           WHEN ANY(candLang IN candidateLanguages WHERE candLang.name = jobLang.name) 
           THEN 
             CASE jobLang.mastery
               WHEN 'native' THEN 10
               WHEN 'fluent' THEN 8
               WHEN 'intermediate' THEN 6
               WHEN 'basic' THEN 4
               ELSE 5
             END
           ELSE 0
         END) * 1.0 / SIZE(jobLanguages)
     END AS languageScore,
     
     // 3. Seniority Level Match (Weight: 20%)
     CASE 
       WHEN job.seniority_level IS NULL THEN 7 // Neutral score if no seniority specified
       WHEN job.seniority_level = candidate.seniority_level THEN 10
       WHEN (job.seniority_level = 'senior' AND candidate.seniority_level = 'mid') OR
            (job.seniority_level = 'mid' AND candidate.seniority_level = 'senior') THEN 8
       WHEN (job.seniority_level = 'mid' AND candidate.seniority_level = 'junior') OR
            (job.seniority_level = 'junior' AND candidate.seniority_level = 'mid') THEN 6
       ELSE 3
     END AS seniorityScore,
     
     // 4. Experience Relevance (Weight: 15%)
     CASE 
       WHEN SIZE(experiences) = 0 THEN 0
       ELSE REDUCE(score = 0, exp IN experiences | 
         score + CASE 
           WHEN exp.company_name = job.company_name THEN 8 // Same company bonus
           WHEN exp.role CONTAINS job.title OR job.title CONTAINS exp.role THEN 6
           WHEN exp.industry = job.industry_sector THEN 4
           ELSE 2
         END) * 1.0 / SIZE(experiences)
     END AS experienceScore,
     
     // 5. Education Match (Weight: 10%)
     CASE 
       WHEN SIZE(education) = 0 THEN 5 // Neutral if no education data
       ELSE REDUCE(score = 0, edu IN education | 
         score + CASE edu.degree
           WHEN 'PhD' THEN 10
           WHEN 'Masters' THEN 8
           WHEN 'Bachelors' THEN 6
           WHEN 'Associates' THEN 4
           ELSE 3
         END) * 1.0 / SIZE(education)
     END AS educationScore,
     
     // 6. Location Preference (Weight: 10%)
     CASE 
       WHEN candidate.location = job.location THEN 10
       WHEN candidate.remote_availability = true AND job.work_mode = 'remote' THEN 9
       WHEN candidate.remote_availability = true AND job.work_mode = 'hybrid' THEN 7
       WHEN candidate.location CONTAINS job.location OR job.location CONTAINS candidate.location THEN 6
       ELSE 3
     END AS locationScore

// Calculate final weighted score
WITH job, candidate,
     skillsScore, languageScore, seniorityScore, experienceScore, educationScore, locationScore,
     
     // Weighted final score calculation
     (skillsScore * 0.35 + 
      languageScore * 0.10 + 
      seniorityScore * 0.20 + 
      experienceScore * 0.15 + 
      educationScore * 0.10 + 
      locationScore * 0.10) AS finalScore,
     
     // Additional factors for tie-breaking
     CASE 
       WHEN candidate.availability = 'immediate' THEN 2
       WHEN candidate.availability = 'two_weeks' THEN 1
       ELSE 0
     END AS availabilityBonus,
     
     // Salary expectation alignment
     CASE 
       WHEN job.salary_min IS NOT NULL AND job.salary_max IS NOT NULL 
            AND candidate.salary_expectation IS NOT NULL THEN
         CASE 
           WHEN candidate.salary_expectation >= job.salary_min 
                AND candidate.salary_expectation <= job.salary_max THEN 2
           WHEN candidate.salary_expectation < job.salary_min THEN 1
           ELSE -1
         END
       ELSE 0
     END AS salaryAlignmentBonus

// Final ranking with all factors
WITH candidate, 
     finalScore + availabilityBonus + salaryAlignmentBonus AS totalScore,
     finalScore, skillsScore, languageScore, seniorityScore, 
     experienceScore, educationScore, locationScore,
     availabilityBonus, salaryAlignmentBonus

// Return ranked candidates with detailed scoring breakdown
RETURN 
  candidate.name AS candidateName,
  candidate.email AS candidateEmail,
  candidate.location AS candidateLocation,
  candidate.seniority_level AS candidateSeniority,
  candidate.availability AS candidateAvailability,
  candidate.salary_expectation AS salaryExpectation,
  
  // Scoring breakdown
  ROUND(totalScore * 100) / 100 AS totalScore,
  ROUND(finalScore * 100) / 100 AS baseScore,
  ROUND(skillsScore * 100) / 100 AS skillsScore,
  ROUND(languageScore * 100) / 100 AS languageScore,
  ROUND(seniorityScore * 100) / 100 AS seniorityScore,
  ROUND(experienceScore * 100) / 100 AS experienceScore,
  ROUND(educationScore * 100) / 100 AS educationScore,
  ROUND(locationScore * 100) / 100 AS locationScore,
  availabilityBonus,
  salaryAlignmentBonus

ORDER BY totalScore DESC, baseScore DESC, skillsScore DESC
LIMIT 50;
""")
```
